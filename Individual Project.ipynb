#### Data Cleaning and Preprocessing

* Rename the column names, as they are too long. 
* For the question: Reasons sleeping in separates bads, our answers loged in separate column for each choice.That's why we would create columns for each reason choice and if the user select that choice we will have that answer if not we will have 0.



```python
#Renaming the columns


data = data.rename({'Which of the following best describes your current relationship status?':'Relationship status',
            'How long have you been in your current relationship? If you are not currently in a relationship, please answer according to your last relationship.':'How long have you been in your current or last relationship',
            "When you're not sleeping in the same bed as your partner, where do you typically sleep?":'If not sleeping in the same bad, where do you sleep?',
            "When you're not sleeping in the same bed, where does your partner typically sleep?":'If not sleeping in the same bad, where your partner sleep',
            'What are the reasons that you sleep in separate beds? Please select all that apply.':'Reasons sleepin in separates bad',
            'To what extent do you agree with the following statement: "sleeping in separate beds helps us to stay together."':'sleeping in separate beds helps us to stay together',
            'To what extent do you agree with the following statement: "we sleep better when we sleep in separate beds."':'we sleep better when we sleep in separate beds.'}, axis = 1)
```

```python
#Renameing the choice columns into reason's name and the user will het the reason's name if user selected that choice 
#and 0 if not.

data.iloc[1:10,9:21]
```
