### Assignmen 1

First of all lets plot the stacked barplot with percenteges. For that we pick two variables and group data by them. My second variable is Age. 

```python
filter_data = ['Age','RespondentID','If you had a smartwatch (like the soon to be released Apple Watch), how likely or unlikely would you be to check the weather on that device?']

data = data[filter_data]
data 
```

```python
#After filtering our data, we need to make a pivot tabel. As we can see we have Age ranges and the numbers of the answers for each answer type.


data_new = pd.pivot_table(data, index='Age', columns=['If you had a smartwatch (like the soon to be released Apple Watch), how likely or unlikely would you be to check the weather on that device?'], aggfunc='count',values='RespondentID')
data_new
```

```python
data_new_1 = data_new.reset_index()
data_new_1
```

```python
# Only numbers can't help us in plotting percentage stacked bar, thats why we do the same thing as above, but also
#sum all numbers for each age range. After that we devide each number by the sum and * 100. So we get the percentage 
#of particular answer:Ro example  "Very likely", in whole "18-29" age range answers.

ax_0= pd.crosstab(data['Age'], data['If you had a smartwatch (like the soon to be released Apple Watch), how likely or unlikely would you be to check the weather on that device?']).apply(lambda r: r/r.sum()*100, axis=1)
ax_1 = ax_0.plot(figsize=(10,10),stacked=True, rot=0,kind = 'barh')

display(ax)

plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc='lower left',
           ncol=4, mode="expand", borderaxespad=0.)
plt.xlabel('Answers number')
plt.ylabel('Age range')

for rec in ax_1.patches:
    height = rec.get_height()
    width = rec.get_width()
    ax_1.text(rec.get_x() + rec.get_width() / 2, 
              rec.get_y() + height / 2,
              "{:.0f}%".format(width),
              ha='center', 
              va='bottom')
```
